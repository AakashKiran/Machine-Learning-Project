```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set(style="whitegrid")

# List numeric and categorical columns
numeric_cols = data.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = data.select_dtypes(include=['object', 'category']).columns

# 1 Distribution plots for numeric features
for col in numeric_cols:
    plt.figure(figsize=(10,4))
    plt.subplot(1,2,1)
    sns.histplot(data[col], kde=True, bins=30)
    plt.title(f'Distribution of {col}')

    plt.subplot(1,2,2)
    sns.boxplot(x=data[col])
    plt.title(f'Boxplot of {col} (Outliers)')

    plt.tight_layout()
    plt.show()

# 2 Count plots for categorical features
for col in categorical_cols:
    plt.figure(figsize=(12,4))
    sns.countplot(x=data[col], order=data[col].value_counts().index)
    plt.xticks(rotation=90)
    plt.title(f'Countplot of {col}')
    plt.show()
```

```python
# Columns to transform
diag_cols = ['diag_1', 'diag_2', 'diag_3']

# First, convert everything to string and replace special/missing codes
for col in diag_cols:
    data[col] = data[col].astype(str)
    data[col] = data[col].replace('?', '0')           # missing → 0
    data[col] = data[col].str.replace('V', '0', regex=False)
    data[col] = data[col].str.replace('E', '0', regex=False)
    data[col] = data[col].astype(float)               # convert to float

# Define ICD-9 ranges and categories
icd_mapping = [
    (1, 140, 1), (140, 240, 2), (240, 280, 3), (280, 290, 4), (290, 320, 5),
    (320, 390, 6), (390, 460, 7), (460, 520, 8), (520, 580, 9), (580, 630, 10),
    (630, 680, 11), (680, 710, 12), (710, 740, 13), (740, 760, 14),
    (760, 780, 15), (780, 800, 16), (800, 1000, 17)
]

# Apply mapping safely using loc
for col in diag_cols:
    for lower, upper, cat in icd_mapping:
        mask = (data[col] >= lower) & (data[col] < upper)
        data.loc[mask, col] = cat
    # Missing values (0) remain 0


```

```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import VarianceThreshold, SelectKBest, f_classif, mutual_info_classif
from sklearn.preprocessing import LabelEncoder


# Separate features and target
X = data.drop('readmitted', axis=1)
y = data['readmitted']









```

```python
'''#  Variance Threshold (remove almost-constant features)
var_thresh = VarianceThreshold(threshold=0.01)
X = pd.DataFrame(var_thresh.fit_transform(X), columns=X.columns[var_thresh.get_support()])
X'''
```

```python
#  Correlation filter (remove highly correlated features)
corr_matrix = X.corr().abs()
print(corr_matrix)
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
to_drop = [col for col in upper.columns if any(upper[col] > 0.85)]
print(to_drop)
X.drop(columns=to_drop, inplace=True)
X
```

```python

#  Mutual Information (keep informative features)
mi = mutual_info_classif(X, y)
mi_scores = pd.Series(mi, index=X.columns)
import matplotlib.pyplot as plt
mi_scores.sort_values().plot(kind='barh', figsize=(8,12))
plt.show()
```
